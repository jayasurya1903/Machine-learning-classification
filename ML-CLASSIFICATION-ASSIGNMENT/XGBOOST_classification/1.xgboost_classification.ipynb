```python
import pandas as pd # import library 
import numpy as np
import lightgbm as lgb
```

```python
dataset= pd.read_csv("CKD.csv")  # function of pandas used to load the data
dataset                        
```

```python
dataset=pd.get_dummies(dataset) #to create dummies value for categorical text
dataset
```

```python
dataset=dataset.replace({True:1,False:0}) #replce the 0 and 1 for true,false
dataset
```

```python
dataset = dataset.drop(columns=['cad_no', 'classification_no', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_a', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_abnormal',
       'rbc_normal', 'pc_abnormal', 'pc_normal', 'pcc_notpresent',
       'pcc_present', 'ba_notpresent', 'ba_present', 'htn_no', 'htn_yes'])# drop columns from the dataset
dataset
```

```python
independent=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 
       'dm_no', 'dm_yes', 'cad_yes', 'appet_poor', 'appet_yes', 'pe_poor',
       'pe_yes', 'ane_no', 'ane_yes']]  # split input and output to get input from dataset 
independent
```

```python
dependent=dataset[['classification_yes']] # get output from dataset
dependent
```

```python
from sklearn.model_selection import train_test_split  # sklearn library for machine learning
x_train,x_test,y_train,y_test = train_test_split(independent, dependent,test_size=0.30,random_state=0)
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV # built model for grid search and cross validation
param_grid=param_grid = {
    'n_estimators': [50, 100, 200],           
    'learning_rate': [0.01, 0.1, 0.3],        
    'max_depth': [3, 5, 7],                        
    'gamma': [0, 1, 5]}

grid= GridSearchCV(XGBClassifier(),param_grid, refit=True, verbose=3, n_jobs=-1)
grid.fit(x_train,y_train)
```

```python
re=grid.cv_results_ # print best parameter after tuning
print("The R_score value for best parameter {}:".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
table
```

```python
grid_predictions = grid.predict(x_test) 
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, grid_predictions)
```

```python
print(cm)
```

```python
# print classification report 
from sklearn.metrics import classification_report
clf_report = classification_report(y_test, grid_predictions)
```

```python
print(clf_report)
```

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(y_test,grid.predict_proba(x_test)[:,1])
```
